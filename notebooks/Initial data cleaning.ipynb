### load the file as Data Frames

```python
files = {
    'DimCity': 'DimCity.csv',
    'DimCustomer': 'DimCustomer.csv',
    'DimDate': 'DimDate.csv',
    'DimEmployee': 'DimEmployee.xlsx',
    'DimStockItem': 'DimStockItem.csv',
    'FactSale': 'FactSale.csv'
}

dimcity = pd.read_csv(files['DimCity'])
dimcustomer = pd.read_csv(files['DimCustomer'], thousands=',')
dimedate = pd.read_csv(files['DimDate'])
dimemployee = pd.read_excel(files['DimEmployee'])
dimstockitem = pd.read_csv(files['DimStockItem'])
factsale = pd.read_csv(files['FactSale'])
```

### dimcustomer data

```python
dimcustomer.columns, dimcustomer.shape
```

```python
dimcustomer.head()
```

###### We have problem with the headers name & the first row

###### clean the headers

```python
dimcustomer.columns = dimcustomer.iloc[0]
dimcustomer = dimcustomer[1:].reset_index(drop=True)
```

```python
dimcustomer.columns
```

```python
dimcustomer.head(2)
```

###### clean Credit Limit column

```python
len(dimcustomer.loc[:,'Credit Limit']) - len(dimcustomer.loc[:,'Credit Limit'].str.startswith('? '))
```

```python
dimcustomer['Credit Limit'] = dimcustomer['Credit Limit'].str[2:]
```

```python
len(dimcustomer.loc[:,'Credit Limit']) - len(dimcustomer.loc[:,'Credit Limit'].str.startswith('.000000'))
```

```python
dimcustomer['Credit Limit'] = dimcustomer['Credit Limit'].str.replace('.000000', '')
dimcustomer['Credit Limit'] = dimcustomer['Credit Limit'].str.replace(',', '')
dimcustomer['Credit Limit'] = dimcustomer['Credit Limit'].str.replace('-', '0')
```

```python
dimcustomer['Credit Limit'] = dimcustomer['Credit Limit'].astype(float)
dimcustomer['Credit Limit'].dtype
```

###### clean the date columns

```python
dimcustomer.head()
```

```python
dimcustomer[['Valid From']].head()
```

```python
dimcustomer['Valid From'] = pd.to_datetime(dimcustomer['Valid From'], errors='coerce')
```

```python
dimcustomer[['Valid From']].head(10)
```

```python
dimcustomer['Valid To'] = pd.to_datetime(dimcustomer['Valid To'], errors = 'coerce')
```

```python
len(dimcustomer['Valid To']) - dimcustomer['Valid To'].isna().sum()
```

```python
dimcustomer['Valid To'].fillna(pd.Timestamp('2099-12-31'), inplace=True)
```

```python
dimcustomer[['Valid To']].head(3)
```

### DimStockItem data

```python
dimstockitem.columns, dimstockitem.shape
```

##### clean headers

```python
dimstockitem.columns = dimstockitem.iloc[0]
dimstockitem = dimstockitem.reset_index(drop=True)
```
